```python
import re

# Path to your log file
log_file = "evaluate_models.log"

print("Reading DOPE score log file...\n")

models = []

with open(log_file, "r") as f:
    for line in f:
        if "DOPE score" in line or "DOPE" in line:
            print(line.strip())
        
        # Extract model name and DOPE score
        match = re.search(r'(\S+\.pdb).*?(-?\d+\.\d+)', line)
        if match:
            model_name = match.group(1)
            dope_score = match.group(2)
            models.append((model_name, dope_score))

print("\n------ Extracted DOPE Scores ------")

for model, score in models:
    print(f"Model: {model}  |  DOPE Score: {score}")
```
